4. Predict "y=income" as a linear function of "x1 = Years of Education" and "x2 = Seniority" for a new individual with (x1 = 18, x2 = 60)

```python
from sklearn.linear_model import LinearRegression
multivariable_lm = LinearRegression()
multivariable_lm.fit(df[['Education', 'Seniority']],df.Income )
prediction = np.array([18,60])
multivariable_lm.predict(prediction.reshape(1,2))
# Predicted income for 18 years of education and 60 seniority is 66.4056967 income
```
